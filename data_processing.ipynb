```python
import pandas as pd
```

```python
df = pd.read_csv('Veterans_Example_Data.csv')
```

```python
# Split the first column into 'State' and 'County'
df[['State', 'County']] = df['Unnamed: 0'].str.split(' -- ', expand=True)

# Remove everything between parentheses (including parentheses) in the 'County' column
df['County'] = df['County'].str.replace(r'\s*\(.*?\)', '', regex=True)

# Drop the original 'Unnamed: 0' column if no longer needed
df.drop(columns=['Unnamed: 0'], inplace=True)

# Display the updated dataframe
# Reorder columns so that 'State' and 'County' are the first two columns
df = df[['State', 'County'] + [col for col in df.columns if col not in ['State', 'County']]]

# Display the updated dataframe
df.head()
```

```python
df['County'] = df['County'].str.title()
df.head()
```

```python
df.to_csv('output.csv', index=False)
```
